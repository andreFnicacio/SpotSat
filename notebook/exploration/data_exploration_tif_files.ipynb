```python
import numpy as np
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

```python
# Carregar e visualizar as bandas de cor
with rasterio.open('/home/andre/Documentos/SpotSat/data/raw/tif_files/red.tif') as src_red:
    red = src_red.read(1)
    show(red, title="Red Band")
```

```python
with rasterio.open('/home/andre/Documentos/SpotSat/data/raw/tif_files/red.tif') as src:
    print(src.meta)  # Imprime os metadados do arquivo .TIF
```
